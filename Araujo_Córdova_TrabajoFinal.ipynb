En el presente trabajo se ha decidido utilizar algunas bases de datos del INEI respecto a la situación de los egresados universitarios. Para ello, en primer lugar, hemos importado las bases de datos como se muestra a continuación:

```python
import pandas as pd
import altair as alt
data1= pd.read_spss("04_EGRESADOS_CAP300_1.sav")
data1.shape
```

```python
data2=pd.read_spss("06_EGRESADOS_CAP400_1.sav")
data2.head()
data2.shape
```

```python
data3=pd.read_spss("08_EGRESADOS_CAP600_1.sav")
data3.head()
data3.shape
```

Con las bases ya importadas, procedemos a realizar el respectivo merge para poder unir las bases de datos.

```python
resultado_merge1 = pd.merge(data1,
                     data2,
                    left_on = 'SELECT',
                    right_on = 'SELECT',
                    how = 'inner')
resultado_merge1
```

```python
resultado_merge_final = pd.merge(resultado_merge1,
                     data3,
                    left_on = 'SELECT',
                    right_on = 'SELECT',
                    how = 'inner', indicator = True)
resultado_merge_final
```

```python
resultado_merge_final.info()
```

En la variable que se muestra a continuación ("CAP600P609_8") se decidió cambiar los valores de números a valores nominales para una mejor comprensión

```python
resultado_merge_final.replace({'CAP600P609_8':{1.0:'si',2.0:'no',3:'no sabe'}},inplace=True)
```

Debido a que Altair solo acepta un máximo de 5000 filas, se procedió a eliminar la cantidad de datos necesarios para poder realizar los gráficos, y fue de manera aleatoria porque no hubo un criterio para eliminar determinadas filas en específico, así que solo nos quedamos con las primeras 5000 unidades.

```python
dtaprev= resultado_merge_final.iloc[0:5000]
dtaprev
```

Ahora, la primera relación que intentamos obtener es sobre una posible influencia del alto costo como motivo para no recomendar la universidad donde estudió con la cantidad de semestres dejados cuando estudiaba.

```python
alt.Chart(dtaprev,  height=500, width=800).mark_bar().encode(
    alt.Y("CAP300P317:Q",  title = "Semestres dejados"),
    alt.X("CAP600P609_8:N",  title = "¿Es el alto costo el principal motivo para no recomendar su universidad?"),tooltip = ["CAP300P317", "CAP600P609_8"]
).interactive().properties(
    title={
      "text": ["¿El alto costo de la universidad influye en la cantidad de semestres dejados?"], 
      "subtitle": ["A través de gráfico de barras",
                   "Fuente: INEI"],
      "color": "Black",
      "subtitleColor": "Green"
    })
```

En base al gráfico de barras obtenido, podemos ver que no existe una correlación entre las variables ya que los individuos que respondieron que el alto costo era un motivo para no recomendar su universidad no influye en la cantidad de semestres que abandonaron.

En segundo lugar, la siguiente relación de variables a experimentar es la de una posible influencia en las veces que repitió secundaria con el sueldo neto mensual que percibe el encuestado.

```python
alt.Chart(dtaprev,  height=500, width=800).mark_bar().encode(
    alt.Y("CAP400P436_MONE:Q",  title = "Sueldo neto mensual"),
    alt.X("CAP300P307:Q",  title = "Veces que repitio secundaria"),tooltip = ["CAP400P436_MONE", "CAP300P307"]
).interactive().properties(
    title={
      "text": ["¿Influye la cantidad de veces que repitio la secundaria en el sueldo neto percibido de los egresados?"], 
      "subtitle": ["A través de gráfico de barras",
                   "Fuente: INEI"],
      "color": "Black",
      "subtitleColor": "Green"
    })
```

En base al gráfico obtenido, podemos sospechar que las personas que no repitieron la secundaria tienen un techo de ingresos más elevado en comparación a aquellos que si repitieron algún grado de secundaria. 

Finalmente, la última relación a probar es una posible influencia de la ayuda de la universidad en conseguir empleo con la cantidad de prácticas realizadas por el egresado.

```python
dtaprev.replace({'CAP400P471':{1.0:'si',2.0:'no',3:'no sabe'}},inplace=True)
alt.Chart(dtaprev,  height=500, width=800).mark_bar().encode(
    alt.Y("CAP300P351:Q",  title = "Número de prácticas realizadas"),
    alt.X("CAP400P471:N",  title = "¿Considera que la universidad lo ayudó a conseguir su primer empleo?"),tooltip = ["CAP300P351", "CAP400P471"]
).interactive().properties(
    title={
      "text": ["¿Influye la ayuda de la universidad para conseguir empleo en la cantidad de practicas realizadas?"], 
      "subtitle": ["A través de gráfico de barras",
                   "Fuente: INEI"],
      "color": "Black",
      "subtitleColor": "Green"
    })
```

En base al gráfico obtenido, podríamos afirmar que aquellos egresados que obtuvieron ayuda por parte de la universidad para obtener su primer empleo realizaron una mayor cantidad de prácticas preprofesionales que aquellos que no recibieron tal ayuda.
